```python
import numpy as np 
import pandas as pd
```

```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx'
```

```python
df_can = pd.read_excel(
    URL,
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

df_can.head()
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.set_index('Country', inplace=True)

df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1)

df_can.head()
```

```python
print("Dimensão dos dados:", df_can.shape)
```

```python
years = list(map(str, range(1980, 2014)))

years
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
# Area Plots
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

#Pèga as top 5 entradas
df_top5 = df_can.head()

df_top5 = df_top5[years].transpose()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10))

plt.title('Trend de imigrantes dos paises top 5')
plt.ylabel('Número de imigrantes')
plt.xlabel('Anos')

plt.show()
```

```python
df_top5.plot(kind='area', 
             alpha=0.25,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

plt.title('Trend de imigrantes dos paises top 5')
plt.ylabel('Número de imigrantes')
plt.xlabel('Anos')

plt.show()
```

```python
# Outra forma de plotagem. Mais flexível
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Trend de imigrantes dos paises top 5')
ax.set_ylabel('Número de imigrantes')
ax.set_xlabel('Anos')
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.index = df_least5.index.map(int)
df_least5.plot(kind='area', alpha=0.45, figsize=(20, 10)) 

plt.title('Trend de imigrantes dos paises top 5 com menos contribuição para imigração')
plt.ylabel('Número de imigrantes')
plt.xlabel('Anos')

plt.show()
```

```python
ax = df_least5.plot(kind='area', alpha=0.55, stacked=False, figsize=(20, 10))
    
ax.set_title('Trend de imigrantes dos paises top 5 com menos contribuição para imigração')
ax.set_ylabel('Número de imigrantes')
ax.set_xlabel('Anos')
```

```python
# Pergunta: Qual é a distribuição de imigração para Dinamarca, Noruega e Suécia nos anos de 1980 a 2013?
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years].plot.hist()
```

```python
# Precisamos fazer o transpose
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()

df_t.plot(kind='hist', figsize=(10, 6))

plt.title('Histograma de imigração para Dinamarca, Noruega e Suécia nos anos de 1980 a 2013')
plt.ylabel('Número de Anos')
plt.xlabel('Número de Imigrantes')

plt.show()
```

```python
# Aumemtando o impacto e estética do plot
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind='hist',
         figsize=(10,6),
         bins = 15,
         alpha=0.6,
         xticks=bin_edges,
         color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histograma de imigração para Dinamarca, Noruega e Suécia nos anos de 1980 a 2013')
plt.ylabel('Número de Anos')
plt.xlabel('Número de Imigrantes')

plt.show()
```

```python
# Impedindo os plots de "stackarem" um sobe o outro
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   
xmax = bin_edges[-1] + 10

# Histograma estacado
df_t.plot(kind='hist',
          figsize=(10, 6), 
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )
plt.title('Histograma de imigração para Dinamarca, Noruega e Suécia nos anos de 1980 a 2013')
plt.ylabel('Número de Anos')
plt.xlabel('Número de Imigrantes')

plt.show()
```

```python
# Pergunta: Use a camada de script para exibir a distribuição de imigração para Grécia, Albânia e Bulgária 
# para os anos de 1980 a 2013? Use uma plotagem sobreposta com 15 compartimentos e um valor de transparência de 0,35.

df_cof = df_can.loc[['Greece', "Albania", "Bulgaria"], years]

#Transpose
df_cof = df_cof.transpose()

#Bins
count, bins_edges = np.histogram(df_cof, 15)

# Histograma
df_cof.plot(kind ='hist',
            figsize=(10, 6),
            bins=15,
            alpha=0.35,
            xticks=bin_edges,
            color=['coral', 'darkslateblue', 'mediumseagreen']
            )
plt.title('Distribuição de imigração para Grécia, Albânia e Bulgária para os anos de 1980 a 2013')
plt.ylabel('Número de Anos')
plt.xlabel('Número de Imigrantes')

plt.show()
```

```python
# Pergunta: Usando o script posterior e o conjunto de dados df_can, 
# crie um gráfico de barra horizontal mostrando o número total de imigrantes para o Canadá dos 15 principais países,
# no período de 1980 a 2013. Rotule cada país com a contagem total de imigrantes.

df_can.sort_values(by='Total', ascending = True, inplace=True)
df_top15 = df_can['Total'].tail(15)
```

```python
# Plot dos dados
df_top15.plot(kind='bar', figsize=(12, 12), color='steelblue')
plt.xlabel('Número de imigrantes')
plt.title("Top 15 paises contribuindo para a imigração para o Canada entre 1980 - 2013")

for index, value in enumerate(df_top15):
    label = format(int(value), ',')
    plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')

plt.show()
```
