```python
# Line Plots
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int) #Muda os valores do index de haiti para tipo inteiro
haiti.plot(kind='line')

plt.title('Imigração do Haiti')
plt.ylabel('Número de imigrantes')
plt.xlabel('Anos')

plt.show()
```

```python
haiti.plot(kind='line')

plt.title("Imigrantes do Haiti")
plt.ylabel("Número de imigrantes")
plt.xlabel("Anos")

plt.text(2000, 6000, 'Terremotos de 2010')

plt.show()
```

```python
df_CI = df_can.loc[['India', 'China'], years]
df_CI
```

```python
df_CI.plot(kind='line')
```

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int) 
df_CI.plot(kind='line')

plt.title('Imigrantes da China e India')
plt.ylabel('Número de imigrantes')
plt.xlabel('Anos')

plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose()

#print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14, 8))

plt.title('Imigrantes dos top 5 países')
plt.ylabel('Número de imigrantes')
plt.xlabel('Anos')


plt.show()
```
